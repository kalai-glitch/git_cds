**While Loop**

```python
# we want to execute the statement repeatedly  then we need a loop
#for loop
#while loop
#syntex
#while condition:
    #statements


# to display the no  from 1 to 10
x=1
while x<=10:
     print(x)  # 2 statement will exceuted whenthe condition is true
     x+=1  ## iteration
print('end')
```

```python
## see the difference from above output with same programe
x=1
while x<=10:
     print(x)  # 2 statement will exceuted whenthe condition is true
     x+=1  ## iteration
     print('end') ## now this in same group statement
```

```python
# program to display even number between m and n
m,n =[int(i) for i in input("enter min and max range:").split(',')]
x=m # start from m onwards
if x%2!=0:
    x=x+1

while x>=m and x<=n:
    print(x)
    x=x+2

```

```python
for i in range(m, n):
    if i%2==0:
        print(i)

#Or just...

print([i for i in range(m, n) if i%2==0])
```

**For loop **
# for loop is useful to iterate over the elements of a sequence. for loop  used to execute a group of statements repeatedly depending  upon the no of elements in sequence. sequence like string, list, tuple,range,arrays etc..

```python
#syntax
#for var in sequence:
    #statements

var='string'
for i in var:
    print(i)
```

```python
#program to display each character from a string using sequence index

str='my world'
n=len(str)
for i in range(n):   #index range 0 to n-1
     print(i)
```

```python
str='my world'
n=len(str)
for i in range(n):   #index range 0 to n-1
     print(str[i])
```

```python
str='my world'
for ch in str:
     print(ch)
```

```python
# to display odd number between 1 and 10
for i in range(1,10,2):
     print(i)
```

```python
# to display numbers in decending order
for x in range(10,0,-1):
      print(x)
```

```python
# program to display and find the sum of a list of numbers
#take list pf number
list=[10,20,30,40,50]
sum=(list[0]+list[1]+list[2]+list[3]+list[4])
print("sum is", sum)
```

```python
# program to display and find the sum of a list of numbers  using for loop
list=[10,20,30,40,50]
sum=0 # initially sum is 0
for i in list:
    print(i)
    sum=sum+i
print('sum=', sum)
```
